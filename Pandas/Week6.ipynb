## Concat, join, merge

```python
ca_youtube = pd.read_csv('./CAvideos.csv')
ca_youtube.head(2)
```

```python
us_youtube = pd.read_csv('USvideos.csv')
us_youtube.head(2)
```

```python
pd.concat([ca_youtube, us_youtube])
```

```python
left = ca_youtube.set_index(['title','trending_date'])
right = us_youtube.set_index(['title','trending_date'])
```

```python
left.head(2)
```

```python
left.join(right, lsuffix='_CAN', rsuffix='_US')
```

```python
print(len(left))
print(len(right))
print(len(left.join(right,lsuffix='_CAN',rsuffix='_US')))
```

```python
#help(left.join)
```

```python
ca_youtube.head(2)
```

```python
ca_channel = ca_youtube.groupby('channel_title').size().sort_values(ascending=False)
```

```python
us_channel = us_youtube.groupby('channel_title').size().sort_values(ascending=False)
```

```python
type(ca_channel)
```

```python
ca_channel.head(2)
```

```python
ca_channel_count = pd.DataFrame({'count':ca_channel})
us_channel_count = pd.DataFrame({'count':us_channel})
print(ca_channel_count.size)
print(us_channel_count.size)
```

```python
ca_channel_count.head(2)
ca_channel_count.merge(us_channel_count, how='inner',left_index=True,
                      right_index=True, suffixes=('_CA','_US'), sort=True)
```
