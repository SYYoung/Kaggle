```python
import pandas as pd
import matplotlib.pyplot as plt
pd.plotting.register_matplotlib_converters()
%matplotlib inline
import seaborn as sns
```

```python
delay_data = pd.read_csv('./Data/flight_delays.csv', index_col='Month')
delay_data.head()
```

```python
airline = delay_data.columns.tolist()
airline
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(data=delay_data[airline[:6]])
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x=delay_data.columns, y=delay_data.loc[1])
```

```python
for i in delay_data.index:
    plt.figure(figsize=(12,6))
    sns.barplot(x=delay_data.columns, y=delay_data.loc[i])
```

```python
max_delay = []
min_delay = []
for i in delay_data.index:
    max_delay.append(delay_data.loc[i].idxmax())
    min_delay.append(delay_data.loc[i].idxmin())
    #print("Max delay airline is: {}, Min delay airline is {}".format(max_delay, min_delay))
print(max_delay, '\n', min_delay)
all_delay = pd.DataFrame({'max delay':max_delay, 'min delay':min_delay}, index=delay_data.index)
all_delay
```

```python
all_delay["max delay"].value_counts()
```
